```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("RTA Dataset.csv")
df=pd.DataFrame(data)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Time']= pd.to_datetime(df['Time'])
```

```python
df.drop(columns=["Time"],inplace=True)
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    mode_values=df[i].mode()[0]
    df[i].fillna(mode_values, inplace=True)
```

```python
df.isnull().sum()
```

```python
sn.boxplot(data=df, y='Number_of_casualties')
plt.show()
```

```python
sn.boxplot(data=df, y='Number_of_vehicles_involved')
plt.show()
```

```python
print(df["Accident_severity"].value_counts())
plt.pie(df["Accident_severity"].value_counts(),labels = df["Accident_severity"].unique())
plt.show()
```

```python
df["Educational_level"].value_counts().plot(kind="bar")
plt.show()
```

```python
print(df["Road_surface_type"].value_counts())
plt.figure(figsize=(10,8))
sn.countplot(x=df["Road_surface_type"], hue=df["Accident_severity"])
plt.xlabel("Road Surface")
plt.show()
```

```python
grid = sn.FacetGrid(data=df, col='Accident_severity', height=4, aspect=1, sharey=False)
grid.map(sn.countplot, 'Number_of_vehicles_involved')
plt.show()
```

```python
lists=['Vehicle_driver_relation', 'Work_of_casuality', 'Fitness_of_casuality','Day_of_week','Casualty_severity','Sex_of_driver',
       'Educational_level','Defect_of_vehicle','Owner_of_vehicle','Service_year_of_vehicle', 'Road_surface_type','Sex_of_casualty','Lanes_or_Medians','Road_allignment',
      'Types_of_Junction','Pedestrian_movement','Age_band_of_casualty']
df.drop(columns = lists, inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
print(df["Light_conditions"].value_counts())
sn.countplot(df["Light_conditions"])
plt.ylabel("Light Conditions")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder           
Le = LabelEncoder()
df=df.apply(Le.fit_transform)  
```

```python
df.head()
```

```python
x=df.drop(columns=["Accident_severity"])
y=df["Accident_severity"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=None)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit(xtrain, ytrain)
```

```python
ypred = clf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import metrics
acc = accuracy_score(ytest, ypred)
print("Logistic Regression model accuracy:", acc*100)
```

```python
print(classification_report(ytest, ypred))
```

```python
plt.figure(figsize=(6, 4))
sn.heatmap(metrics.confusion_matrix(ytest, ypred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(xtrain, ytrain)
```

```python
y1pred = knn_classifier.predict(xtest)
```

```python
accuracy = accuracy_score(ytest, y1pred)
print("KNN Model Accuracy:", accuracy*100)
```

```python
print(classification_report(ytest, y1pred))
```

```python
plt.figure(figsize=(6, 4))
sn.heatmap(metrics.confusion_matrix(ytest, y1pred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(xtrain, ytrain)
```

```python
y2pred = clf.predict(xtest)
accuracy = accuracy_score(ytest, y2pred)
print("Decision Tree Accuracy:", accuracy*100)
```

```python
print(classification_report(ytest, y2pred))
```

```python
plt.figure(figsize=(6, 4))
sn.heatmap(metrics.confusion_matrix(ytest, y2pred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.svm import SVC
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(xtrain, ytrain)
```

```python
y3pred = svm_classifier.predict(xtest)
accuracy = accuracy_score(ytest, y3pred)
print("Decision Tree Accuracy:", accuracy*100)
```

```python
print(classification_report(ytest, y3pred))
```

```python
plt.figure(figsize=(6, 4))
sn.heatmap(metrics.confusion_matrix(ytest, y3pred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
